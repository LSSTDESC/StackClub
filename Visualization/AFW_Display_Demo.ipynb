## Import Common Python Libraries

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astropy.visualization import ZScaleInterval
zscale = ZScaleInterval()
```

## Loading the LSST DM Stack

```python
import lsst.daf.persistence as dafPersist
```

```python
import lsst.afw.display     as afwDisplay
```

```python
datadir = "/project/shared/data/Twinkles_subset/output_data_v2"
butler = dafPersist.Butler(datadir)
```

```python
# Grab a calexp of interest
dataId = {'filter': 'r', 'raft': '2,2', 'sensor': '1,1', 'visit': 235}
calexp = butler.get('calexp', **dataId)
```

```python
# Use lsst.afw.display with the matplotlib backend# Use l 
afwDisplay.setDefaultBackend('matplotlib') 
```

```python
plt.figure()
afw_display = afwDisplay.Display()
afw_display.scale('asinh', 'zscale')
afw_display.mtv(calexp.image)
```

```python
method_list = [func for func in dir(afw_display) if callable(getattr(afw_display, func))]
```

```python
print(method_list)
```

```python
plt.figure()
afw_display = afwDisplay.Display()
afw_display.mtv(calexp.image)
plt.show()
```

```python
print(afw_display.scale().__doc__)
```
